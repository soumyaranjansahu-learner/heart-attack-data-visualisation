```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv('//Users//soumya//Downloads//heart.csv')
```

```python
display(data.head(10))
```

```python
#Scatter plot with age against oldpeak
plt.scatter(data['age'],data['oldpeak'])

#adding title to the plot
plt.title("Scatter plot")

#setting the x and y labels
plt.xlabel('age')
plt.ylabel('oldpeak')

#adding the legends
plt.show()
```

```python
#Line chart with age against oldpeak
plt.plot(data['age'])
plt.plot(data['oldpeak'])

#adding title to the plot
plt.title("Line chart")

#setting the x and y labels
plt.xlabel('age')
plt.ylabel('oldpeak')

#adding the legends
plt.show()
```

```python
#Bar chart or bar plot
plt.bar(data['age'],data['oldpeak'])

#adding title to the plot
plt.title("Bar chart")

#setting the x and y labels
plt.xlabel('age')
plt.ylabel('oldpeak')

#adding the legends
plt.show()
```

```python
#Histogram
plt.hist(data['thalachh'])

plt.title("histogram")

plt.show()
```
